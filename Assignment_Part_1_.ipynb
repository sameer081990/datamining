**Question 3:**

The accuracy of 99.9% for classifier can be terrible for some datasets because accuracy  is not the only measure that needs to be considered. We also need to take into consideration precision and recall. In some cases although the accuracy might be high it so happens that precision or recall or both might be low. For instance when precision is high and recall is low, whatever the classifier predicted is correct but it will be predicting only small amount which is not considered good because we want our classifier to be able to predict a large amount. And in the case where precision is low and recall is high although our classifier predicts a huge amount but the data it predicted might not be right. It can only be considered as a good classifier when all the three measures accuracy precision and recall are high.
